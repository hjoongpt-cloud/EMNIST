```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("../outputs/stage_E_deep/variants_summary.csv")

# : variant test_acc boxplot
df.boxplot(column="test_acc", by="variant")
plt.title("Test Acc by Variant"); plt.suptitle(""); plt.ylabel("acc")
plt.tight_layout(); 
plt.savefig("../outputs/stage_E_deep/testacc_by_variant.png")

# misroute_rate / gate_regret_ce bar  
mean_df = df.groupby("variant").mean(numeric_only=True)
mean_df[["misroute_rate","gate_regret_ce","entropy_mean"]].plot(kind="bar")
plt.tight_layout(); plt.savefig("../outputs/stage_E_deep/gating_metrics_by_variant.png")
```
